```python
import pandas as pd
```

```python
library = pd.read_csv('./csv/library/ulsan_library.csv', encoding='euc-kr')
```

```python
library
```

```python
library.columns
```

```python
library1 = library[['도서관명','소재지도로명주소']]
library1
```

```python
library2 = library1[library1.도서관명 != '도서관명']
library2
```

```python
small_addr = list()

addr = library2.loc[:, ['소재지도로명주소']].values 
for tmp in addr :
    small_addr.append(tmp[0].split(" ")[1])

print(small_addr)
```

```python
library2['GUGUN'] = small_addr
library2
```

```python
library2.drop(['소재지도로명주소'], axis='columns', inplace=True)
```

```python
library3 = library2.groupby('GUGUN').count()
library3
```

```python
library4 = library3.rename(columns={'도서관명':'도서관수'})
library4
```

```python
library4.to_csv('./csv/1_ULSAN/jiyoon/ulsan/ulsan_gugun_library.csv', encoding='euc-kr')
```
