```python
parking = parking.loc[:, ['주차장명','주소']]
parking
```

```python
small_addr = list()

addr = parking.loc[:, ['주소']].values 
for tmp in addr :
    small_addr.append(tmp[0].split(" ")[1])

print(small_addr)
```

```python
parking['GUGUN'] = small_addr
```

```python
df = parking.drop(['주소'], axis='columns', inplace=True)
print(df)
```

```python
parking.head()
```

```python
parking = parking.groupby("GUGUN").count()
```

```python
parking
```

```python
parking.to_csv('./csv/parking/gugun_parking.csv', encoding='euc-kr')
```
