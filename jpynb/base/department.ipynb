```python
import pandas as pd
```

```python
department = pd.read_csv('./csv/department/department.csv', encoding='utf-8')
```

```python
print(department)
```

```python
department
```

```python
department.drop(['Unnamed: 0'], axis='columns', inplace=True)
print(department)
```

```python
li = department['addr'].values
```

```python
gugun = list()
for i in li:
    print(i)
    tmp = i.split(' ')
    print(tmp[1])
    gugun.append(tmp[1])
department['GUGUN'] = gugun 
```

```python
department
```

```python
df = department.drop(['addr'], axis='columns', inplace=True)
```

```python
department
```

```python
df = department.groupby('GUGUN').count()
df
```

```python
df.rename(columns = {'Name':'백화점수'},inplace=True)
df
```

```python
df.to_csv('./csv/department/gugun_department.csv', encoding='euc-kr')
```
