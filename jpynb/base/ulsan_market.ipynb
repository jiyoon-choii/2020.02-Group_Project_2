```python
import pandas as pd
```

```python
market = pd.read_csv('./csv/market/ulsan_market.csv', encoding='euc-kr')
```

```python
market.head()
```

```python
market = market[['시장명','소재지도로명주소']]
market
```

```python
type(market)
```

```python
market1 = market.drop([36,37])
market1
```

```python
small_addr = list()

addr = market1.loc[:, ['소재지도로명주소']].values 
for tmp in addr :
    small_addr.append(tmp[0].split(" ")[1])

print(small_addr)
```

```python
market1['GUGUN'] = small_addr
```

```python
market1
```

```python
market2 = market1.groupby('GUGUN').count()
```

```python
market3 = market2.drop(['시장명'],axis=1)
market3
```

```python
market4 = market3.rename(columns={'소재지도로명주소':'시장수'})
market4
```

```python
market4.to_csv('./csv/1_ULSAN/jiyoon/ulsan/ulsan_gugun_market.csv', encoding='euc-kr')
```
