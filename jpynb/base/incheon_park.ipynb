```python
park1 = park.loc[:, ['공원명','소재지도로명주소']]
park1
```

```python
park2 = park1.fillna(0)
```

```python
park3 = park2[park2.소재지도로명주소 != 0]
park3
```

```python
park3 = park3[park3.공원명 != '공원명']
park3
```

```python
small_addr = list()

addr = park3.loc[:, ['소재지도로명주소']].values 
for tmp in addr :
    small_addr.append(tmp[0].split(" ")[1])

print(small_addr)
```

```python
park3['GUGUN'] = small_addr
park3
```

```python
park3.drop(['소재지도로명주소'], axis='columns', inplace=True)
```

```python
park4 = park3.groupby('GUGUN').count()
```

```python
park5 = park4.rename(columns={'공원명':'공원수'})
park5
```

```python
park5.to_csv('./csv/2_INCHEON/jiyoon/incheon_gugun_park.csv', encoding='euc-kr')
```
