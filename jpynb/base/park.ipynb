```python
park = park.loc[:, ['park_nm','lnm_adres']]
park
```

```python
small_addr = list()

addr = park.loc[:, ['lnm_adres']].values 
for tmp in addr :
    small_addr.append(tmp[0].split(" ")[1])

print(small_addr)
```

```python
park['GUGUN'] = small_addr
```

```python
df = park.drop(['lnm_adres'], axis='columns', inplace=True)
print(df)
```

```python
park.rename(columns = {'park_nm':'공원명'},inplace=True)
park
```

```python
park = park.groupby("GUGUN").count()
```
