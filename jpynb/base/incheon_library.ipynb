```python
import pandas as pd
```

```python
library = pd.read_csv('./csv/library/incheon_library.csv', encoding='euc-kr')
```

```python
library
```

```python
library1 = library[['도서관명','시군구명']]
library1
```

```python
library2 = library1[library1.도서관명 != '도서관명']
library2
```

```python
library3 = library2.rename(columns={'시군구명':'GUGUN'})
library3
```

```python
library4 = library3.groupby('GUGUN').count()
library4
```

```python
library5 = library4.rename(columns={'도서관명':'도서관수'})
library5
```

```python
library5.to_csv('./csv/2_INCHEON/jiyoon/incheon_gugun_library.csv', encoding='euc-kr')
```
