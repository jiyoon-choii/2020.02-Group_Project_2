```python
import pandas as pd
```

```python
busan = pd.read_csv('./csv/BUSAN_ALL.csv', encoding='utf-8')
```

```python
ulsan = pd.read_csv('./csv/ULSAN_ALL.csv', encoding='euc-kr')
```

```python
incheon = pd.read_csv('./csv/INCHEON_ALL.csv', encoding='euc-kr')
```

```python
busan = busan.rename(columns={'Unnamed: 0':'GUGUN'})
busan
```

```python
busan=busan.set_index('GUGUN')
```

```python
busan.index
```

```python
di=dict()

for i in range(len(busan.index)):
    di[busan.index[i]]='부산_'+busan.index[i]
    
di
```

```python
for i in range(len(busan.index)):
    busan.rename(index=di, inplace=True)
busan
```

```python
ulsan=ulsan.set_index('GUGUN')
```

```python
ulsan.index
```

```python
di=dict()

for i in range(len(ulsan.index)):
    di[ulsan.index[i]]='울산_'+ulsan.index[i]
    
di
```

```python
for i in range(len(ulsan.index)):
    ulsan.rename(index=di, inplace=True)
ulsan
```
