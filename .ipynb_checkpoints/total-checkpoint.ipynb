```python
total = pd.read_csv('./csv/total_data.csv', encoding='utf-8')
total = total.rename(columns={'Unnamed: 0':'GUGUN'})
total
total = total.set_index('GUGUN')
total
```

```python
department = pd.read_csv('./csv/department/gugun_department.csv', encoding='euc-kr')
department = department.set_index('GUGUN')
department
```

```python
library = pd.read_csv('./csv/library/gugun_library.csv', encoding='euc-kr')
library = library.set_index('GUGUN')
library = library.iloc[:, [1]]


```

```python
library = library.rename(index={'부산중구':'중구'})
```

```python
library
```

```python
market = pd.read_csv('./csv/market/gugun_market.csv', encoding='euc-kr')
market = market.set_index('GUGUN')
market
market = market.iloc[:, [1]]
market
```

```python
total = pd.concat([department, library, market], axis=1)
total
```

```python
total
```

```python
total = total.fillna(0)
total
```

```python
total = total.rename(columns={'count_bus':'bus_cnt', 'railroad_count':'railroad_cnt', '공원명':'park_cnt', '주차장명':'parking_cnt', '2018_계_시설수 (개)':'sisul', '백화점수':'department_cnt', '도서관수':'library_cnt', '시장수':'market_cnt'})
total
```

```python
real_total = pd.concat([real,total], axis=1)
real_total
```

```python
real_total.to_csv('./csv/real_total.csv', encoding='utf-8' )
```
