```python
import pandas as pd
```

```python
ul = pd.read_csv('./csv/population/ulsan_pop.csv', encoding='euc-kr')
```

```python
ul.drop('구분', axis=1)
ul
```

```python
ul1 = ul[ul.구분 == '총인구수']
ul1
```

```python
ul2 = pd.melt(ul1, id_vars='구분')
ul2
```

```python
ul3 = ul2.rename(columns={'variable':'GUGUN', 'value':'인구수'})
ul3
```

```python
ul3.drop('구분', axis=1, inplace=True)
ul3
```

```python
ul4 = ul3.drop([0])
ul4
```

```python
ul5 = ul4.set_index('GUGUN')
```

```python
ul5
```
